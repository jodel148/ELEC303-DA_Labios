<h1>Exploratory Data Analysis</h1>

<h1> Insight 1: Regional Comparison of Average Household Income</h1>

```python
import matplotlib.pyplot as plt

df.groupby('Region')['Total Household Income'].mean()\
    .sort_values(ascending=False)\
    .plot(kind='bar', figsize=(12,6), title='Average Household Income by Region')


plt.ylabel('Average Household Income')  # Optional: Add y-axis label
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

<h2>This barplot illustrates the variation in average household income across different regions.

NCR stands out with a significantly higher average income, exceeding â‚±400,000, which is noticeably above all other regions. Following that, 4A - Calabarzon have averages around â‚±300,000, while most others range between â‚±200,000 to â‚±270,000.

The lowest average income is just above â‚±150,000, highlighting a wide income gap among regions. This disparity suggests that economic opportunities and living standards may vary greatly depending on location, with wealth likely being concentrated in a few urban or highly developed areas, while others lag behind.</h2>

<h2>Insight 2: Average Meat Expenditure by Income Level</h2>

```python
bins = [0, 100000, 200000, 300000, 400000, 1000000]
labels = ['<100k', '100k-200k', '200k-300k', '300k-400k', '400k+']
df['Income Group'] = pd.cut(df['Total Household Income'], bins=bins, labels=labels)
df.groupby('Income Group')['Meat Expenditure'].mean().plot(kind='bar')



```

<h2><hThis barplot shows the distribution of data across different income groups. As the income group increases, the number of individuals or observations in that group also increases.

The highest income group has the most data points, indicating that more people fall into higher-income brackets. Conversely, the lowest income group has the fewest data points, suggesting fewer people in this category. 

<h1>Insight 3: Distribution of House Floor Area</h1>

```python
sns.histplot(df['House Floor Area'], bins=30, kde=True)
```

<h2>This histogram with a density line shows the distribution of house floor area sizes.
Most houses are small, with the majority having a floor area between 0 and 100 square meters. The highest number of houses (the peak) is found below 50 square meters.
The graph is right-skewed, which means that while most houses are small, there are a few much larger ones.
Beyond 100 square meters, the number of houses quickly drops.
Houses larger than 200 square meters are very rare, and anything close to 1000 square meters is considered an outlier in this data.<h2></h2>
