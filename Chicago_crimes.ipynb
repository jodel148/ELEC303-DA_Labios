<br>
<h3>Fill up null values</h3>

```python
chicago_crimes_df['Location Description'] = chicago_crimes_df['Location Description'].fillna('unaccounted')
chicago_crimes_df['X Coordinate'] = chicago_crimes_df['X Coordinate'].fillna(chicago_crimes_df['X Coordinate'].mode()[0])
chicago_crimes_df['Y Coordinate'] = chicago_crimes_df['Y Coordinate'].fillna(chicago_crimes_df['Y Coordinate'].mode()[0])
chicago_crimes_df['Latitude'] = chicago_crimes_df['Latitude'].fillna(chicago_crimes_df['Latitude'].mode()[0])
chicago_crimes_df['Longitude'] = chicago_crimes_df['Longitude'].fillna(chicago_crimes_df['Longitude'].mode()[0])
chicago_crimes_df['Location'] = chicago_crimes_df['Location'].fillna(chicago_crimes_df['Location'].mode()[0])
```

```python
chicago_crimes_df.isnull().sum()
```

<br>
<h3>Changing data types</h3>

```python
chicago_crimes_df['Case Number'] = chicago_crimes_df['Case Number'].astype('category')
chicago_crimes_df['Date'] = chicago_crimes_df['Date'].astype('str')
chicago_crimes_df['Block'] = chicago_crimes_df['Block'].astype('category')
chicago_crimes_df['IUCR'] = chicago_crimes_df['Block'].astype('category')
chicago_crimes_df['Primary Type'] = chicago_crimes_df['Primary Type'].astype('category')
chicago_crimes_df['Description'] = chicago_crimes_df['Description'].astype('category')
chicago_crimes_df['Location Description'] = chicago_crimes_df['Location Description'].astype('category')
chicago_crimes_df['FBI Code'] = chicago_crimes_df['FBI Code'].astype('category')
chicago_crimes_df['Updated On'] = chicago_crimes_df['Updated On'].astype('str')
chicago_crimes_df['Location'] = chicago_crimes_df['Location'].astype('category')

chicago_crimes_df['Date'] = chicago_crimes_df['Date'].str.strip()
chicago_crimes_df['Date'] = chicago_crimes_df['Date'].astype('str')
chicago_crimes_df['Date'] = chicago_crimes_df['Date'].str.replace('/', '-')
chicago_crimes_df['Date'] = pd.to_datetime(chicago_crimes_df['Date'], dayfirst = True, errors='coerce')
chicago_crimes_df['Year'] = chicago_crimes_df['Date'].dt.year
chicago_crimes_df['Month'] = chicago_crimes_df['Date'].dt.month
chicago_crimes_df['Day']  = chicago_crimes_df['Date'].dt.day
chicago_crimes_df['DayOfWeek']  = chicago_crimes_df['Date'].dt.dayofweek

chicago_crimes_df['Updated On'] = chicago_crimes_df['Updated On'].str.strip()
chicago_crimes_df['Updated On'] = chicago_crimes_df['Updated On'].astype('str')
chicago_crimes_df['Updated On'] = chicago_crimes_df['Updated On'].str.replace('/', '-')
chicago_crimes_df['Updated On'] = pd.to_datetime(chicago_crimes_df['Updated On'], dayfirst = True, errors='coerce')
chicago_crimes_df['Year'] = chicago_crimes_df['Updated On'].dt.year
chicago_crimes_df['Month'] = chicago_crimes_df['Updated On'].dt.month
chicago_crimes_df['Day']  = chicago_crimes_df['Updated On'].dt.day
chicago_crimes_df['DayOfWeek']  = chicago_crimes_df['Updated On'].dt.dayofweek
```
