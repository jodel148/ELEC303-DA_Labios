<h2>Monthly dengue case from 2016 to 2020?</h2>

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
Dengue_df['Month'] = pd.Categorical(Dengue_df['Month'], categories =month_order, ordered = True )
monthly_cases = Dengue_df.groupby(['Month', 'Year'])['Dengue_Cases'].sum()
```

```python

monthly_cases.unstack().plot(kind='line')
plt.gcf().set_size_inches(12,8)
plt.grid(True)
plt.xticks(range(0,12), month_order)
plt.show()
```

<h3>Insigth No. 1: Based on the Data shown above. From month of january to march, june to august and october to december cases of dengue has always increased in this month yearly. But the highest cases of dengue is in the month of july. In 2020, the Philippines recorded a substantial decrease in the number of dengue cases, with a reduction in notified cases of about 70â€“90% during the rainy season . Apart from existing control and prevention measures implemented in the country, such as the establishment of dengue centres of excellence in tertiary hospitals and the creation of dengue fast lanes.The decrease in the number of cases during the COVID-19 pandemic may be largely due to the reduced mobility of the population. </h3>

<h3>Insigth No. 2: Month of june to december is a rainy season which Dengue cases are most prevalent in the Philippines when there is more rainfall, which creates more breeding sites for mosquitoes, and higher temperatures, which accelerate mosquito development.</h3>
